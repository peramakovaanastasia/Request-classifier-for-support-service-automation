```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('correct_dataset.csv')
df.head()
```

Анализ распределения классов

Узнать равномерность количества данных в каждом классе

```python
plt.figure(figsize=(10, 5), facecolor='#FFFACD')
df['category'].value_counts().plot(kind='bar', color='#FFA500')
plt.title('Распределение категорий')
plt.xlabel('Категории')
plt.ylabel('Количество тестов')
plt.show()
```

Длина текстов для каждой категории

```python
df['length'] = df['text'].apply(lambda x: len(str(x).split()))

plt.figure(figsize=(10, 5), facecolor='#FFFACD')
df.groupby('category')['length'].mean().plot(kind='bar', color='#FFA500')
plt.title('Средняя длина текстов по категориям')
plt.xlabel('Категории')
plt.ylabel('Среднее количество слов')
plt.show()
```
